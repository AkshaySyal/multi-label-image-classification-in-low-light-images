```python
import pandas as pd
import numpy as np
```

```python
import os
```

```python
labels = ['Bicycle', 'Boat', 'Bottle', 'Bus', 'Car', 'Cat', 'Chair', 'Cup', 'Dog', 'Motorbike', 'People', 'Table']
```

```python
cols = ['name','Bicycle', 'Boat', 'Bottle', 'Bus', 'Car', 'Cat', 'Chair', 'Cup', 'Dog', 'Motorbike', 'People', 'Table']
```

```python
# Generating labels for train data
```

```python
data_dir = 'C:\\Users\\91882\\Desktop\\PA review 1\\Groundtruth\\ExDark_Annno\\train\\'
```

```python
data = pd.DataFrame(columns=cols)
```

```python
for lbl in labels:
    for img in os.listdir(data_dir+lbl):
        path_to_file = data_dir+lbl+'\\'+img
        file = open(path_to_file, 'r')
        Lines = file.readlines()
        Lines.pop(0)
        row = {'name': img, 'Bicycle':0, 'Boat':0, 'Bottle':0, 'Bus':0, 'Car':0, 'Cat':0, 'Chair':0, 'Cup':0, 'Dog':0, 'Motorbike':0, 'People':0, 'Table':0}  
        for line in Lines:
            obj = line.split(' ')[0]
            if(row[obj]==0):
                row[obj]=row[obj]+1
        data = data.append(row,ignore_index=True)
      
```

```python
data
```

```python
data.to_csv('train.csv',index=False)
```
